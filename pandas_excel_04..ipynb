```python
import pandas as pd
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell
df=pd.read_excel("c:\\datas\\Stock1.xlsx", converters ={
    'people':convert_people_cell,
    'eps': convert_eps_cell
})
df
```

```python
df.to_excel("c:\\datas\\new1.xlsx",sheet_name="stockss")
```
